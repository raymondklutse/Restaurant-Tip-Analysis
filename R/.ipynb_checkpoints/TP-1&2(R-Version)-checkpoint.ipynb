### Question 1

```r
data <- read.csv('../dataset/tips.csv',header=FALSE)
```

```r
#int stands for integers.
#dbl stands for doubles, or real numbers.
#chr stands for character vectors, or strings.
#dttm stands for date-times (a date + a time).

head(data)
```

```r
#list() is used to create lists
#c() is used to create vectors
#names() returns columns of dataframe
names(data) <- c("obs","totbill","tip","sex","smoker","day","time","size")
```

```r
#dim() is used to get the dimensions of the data 
head(data)
```

```r
summary(data)
```

```r
str(data)
```

```r
#change sex to Male & Female
data$sex[data$sex ==0]= "Male"
data$sex[data$sex ==1]= "Female"
```

```r
#change Smoker to Yes or No
data$smoker[data$smoker ==0]= "No"
data$smoker[data$smoker ==1]= "Yes"
```

```r
#change day to Thurs,Frid,Sat & Sun
data$day[data$day==3] = "Thur"
data$day[data$day==4] = "Fri"
data$day[data$day==5] = "Sat"
data$day[data$day==6] = "Sun"
```

```r
#change Smoker to Yes or No
data$time[data$time==0] = "Day"
data$time[data$time==1] = "Night"
```

```r
summary(data)
```

# Question 2

```r
library(dplyr)
library(tibble)
tiprate <- data$tip*  100 /data$totbill
data <- add_column(data, tiprate = tiprate, .after = 3)
qt_data <- select(data,totbill,tip,tiprate,size)
head(qt_data)
```

```r
summary(qt_data)
```

```r
#Quantitative Data
p1 <- ggplot(qt_data, aes(x = totbill, fill = totbill)) + geom_histogram(bins= 20)
p2 <- ggplot(qt_data, aes(x = tip, fill = tip))+ geom_histogram(bins= 20)
p3 <- ggplot(qt_data, aes(x = size, fill = size)) + geom_histogram(bins= 20)
p4 <- ggplot(qt_data, aes(x = tiprate, fill = size)) + geom_histogram(bins= 20)
plot_grid(p1, p2,p3,p4, labels = "AUTO")
```

```r
fri=factor(with(data,ifelse((day=="Frid"),1,0)))
thu=factor(with(data,ifelse((day=="Thur"),1,0)))
sat=factor(with(data,ifelse((day=="Sat"),1,0)))
sun=factor(with(data,ifelse((day=="Sun"),1,0)))
tips2=cbind(data,thu,fri,sat,sun)
str(tips2)
```

```r
regmodel1=lm(tiprate~sex+smoker+time+thu,data=tips2) 
summary(regmodel1)
```
